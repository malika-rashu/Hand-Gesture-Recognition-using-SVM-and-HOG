##loading images from the directory

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        if filename!=".ipynb_checkpoints":
          img = img_to_array(load_img(os.path.join(folder,filename)))
          if img is not None:
            
              images.append(img)
    return images
```

#data augmentation

data augmentation on images with label 0

```python
augmented_images0=[]

count=0
for i in range(len(images_0)):
  
  #img = load_img('./images_0/'+ str(8*count+1)+ '.png')
  data = (images_0[i])
  count+=1
  # expand dimension to one sample
  samples = expand_dims(data, 0)
  
  # create image data augmentation generator
  datagen = ImageDataGenerator(width_shift_range=[-15,15],height_shift_range= [-15,15],horizontal_flip=True, rotation_range=45,zoom_range=[1.25,1.25])
  
  # prepare iterator
  it = datagen.flow(samples, batch_size=1)
  # generate samples and plot
  for i in range(8):
    # define subplot
    
    # generate batch of images
    batch = it.next()

    # convert to unsigned integers for viewing
    imge = batch[0].astype('uint8')
    augmented_images0.append((imge))

    # plot raw pixel data
    #cv2.imwrite(os.path.join(out_path , 'P0_'+str(count)+'_'+str(i)+'.png'), imge)
    if count  == 1:
      pyplot.subplot(330 + 1 + i)
      pyplot.imshow(imge)
# show the figure
#pyplot.show()
```

#applying HOG on canny edge detected images

```python
hog_images = [] 
hog_features = [] 
labels = [] 
```

```python
count=0
for image in augmented_images0:
  count+=1
  image= np.uint8(image)
  image = cv2.Canny(image, 100, 200,3 , L2gradient=True)
  if count==1:
    plt.imshow(image)
  hog_vector, hog_image = hog(image, orientations=9, pixels_per_cell=(16, 16),cells_per_block=(1, 1), visualize=True)
  hog_images.append(hog_image)
  #plt.imshow(hog_image)
  hog_features.append(hog_vector) 
  labels.append(0) 
```

randomly shuffling the data

```python
labels = np.array(labels).reshape(len(labels),1)
data_frame = np.hstack((hog_features,labels))
hog_features = np.array(hog_features)
np.random.shuffle(data_frame)
```

partitioning the data into training set and testing set

```python
partition = int(len(hog_features)*80/100)
```

```python
x_train, x_test = data_frame[:partition,:-1],  data_frame[partition:,:-1]   #extracting the hog fecture vectors
y_train, y_test = data_frame[:partition,-1:].ravel() , data_frame[partition:,-1:].ravel() #extracting the labels
```

#SVM multiclass

fittng the data using four different types of kernels

```python
linear = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo').fit(x_train, y_train)
rbf = svm.SVC(kernel='rbf', gamma=1, C=1, decision_function_shape='ovo').fit(x_train, y_train)
poly = svm.SVC(kernel='poly', degree=4, C=1, decision_function_shape='ovo').fit(x_train, y_train)
sig = svm.SVC(kernel='sigmoid', C=1, decision_function_shape='ovo').fit(x_train, y_train)
```

predicting on the test data

```python
linear_pred = linear.predict(x_test)
poly_pred = poly.predict(x_test)
rbf_pred = rbf.predict(x_test)
sig_pred = sig.predict(x_test)
```

calculating accuracy scores of all the four cases with different kernels

```python
accuracy_lin = linear.score(x_test, y_test)
accuracy_poly = poly.score(x_test, y_test)
accuracy_rbf = rbf.score(x_test, y_test)
accuracy_sig = sig.score(x_test, y_test)
print("Accuracy Linear Kernel", accuracy_lin)
print("Accuracy Polynomial Kernel:", accuracy_poly)
print("Accuracy Radial Basis Kernel:", accuracy_rbf)
print("Accuracy Sigmoid Kernel:", accuracy_sig)
```

detailed analysis of the case of polynomial kernel

```python
from sklearn.metrics import classification_report,accuracy_score

print("Accuracy Percentage: "+str(accuracy_score(y_test, poly_pred)*100)+"% ")
print('\n')
print(classification_report(y_test, poly_pred))
```
